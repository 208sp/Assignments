```python
import os
os.mkdir("Database")
db = open("Database.txt","a")
db.write("Shivam@gmail.com" + " " + "Shivam@123" + os.linesep)
db.close()

def option():
  print("Enter an option: ")
  option = (input("\n option 1: Register \n option 2: Login \n option 3: Forgot password \n "))
  if option == '1':
    return register()
  elif option == '2':
    return check_details()
  elif option == '3':
    return forget_pass()
  else:
    return 'enter a valid option'

def existing_user():
  with open("Database.txt","r") as fp:
    for line in fp.readlines():
      username, password = line.split(" ")
    yield username, password

def validate(username, password):
  return any(i ==(username, password) for i in existing_user())

def user_exists(name):
  for username, password in existing_user():
    if username == name:
      return True
  return False

def retrive(name):
  for username, password in existing_user():
    if username == name:
      return "The retrived password is: " +password

def get_email_password():
  name = email_check()
  password = password_check()
  return name, password

def ask_input():
  name = str(input("email id:"))
  password = input("password:")
  return name, password

def check_details():
  name, password = ask_input()
  if validate(name, password):
    return "Login Success!!!"

def forget_pass():
  name, password = ask_input()
  if user_exists(name):
    return retrive(name)

def register():
  name, password = get_email_password()
  if name == 'Invalid Email id':
    return 'Enter a valid email id'
  if password == 'invalid password':
    return 'Enter a valid password'
  elif user_exists(name):
    return "Name Unavailable. Please try again"
  else:
     db = open("Database.txt","a")
     db.write(name + " " + password + os.linesep)
     print("Registration Success !!!")
     db.close()

def email_check():
  email = input("email_id : ")
  x = list(email)
  if '@' in email and '.' in email:
    diff = x.index('.')-x.index("@")
    d,e = 0,0
    if diff < 1:
      d+=1
    if x[0].isalpha():
      e+=1
    if d == 0 & e == 0:
      return email 
    else:
      print('email id should contain @ and .')
      return 'invalid email id'
  else:
    print('invalid email id')
    return 'invalid email id'

def password_check():
  password = input("password: ")
  special_char =['$', '@', '#','%', '~', '!', '*', '^', '&', '/', '-']
  val=True
  if len(password) <5 and len(password) > 16:
    print('length should be b/w 5 and 16')
    val = False
  if not any(item.isdigit() for item in password):
    print('Password should have at least one digit')
  if not any(item.isupper() for item in password):
    print('Password should have at least one uppercase letter')
    val = False
  if not any(item.islower() for item in password):
    print('Password should have at least one lowercase letter')
    val = False
  if not any(item in special_char for item in password):
    print('password should have at least one special characters')
    val = False
  if val:
    return password
  else:
    return 'invalid password'

option()
```
